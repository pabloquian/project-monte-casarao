### NOTE

From 2007 to 2012 the client calculated the 'cottage_total' like this: ('nights_w' x 'dayprice_w') + ('nights_d' x 'dayprice_d') - 'discount_euro' = 'cottage_total'.

In this period the column 'weeks' means the number of weeks the guest stayed

From 2013 on the client calculated the 'cottage_total' like this: ('weeks' x 'weekprice_w') + ('extra_days' x 'dayprice_d') - 'discount_euro' = ' cottage_total'

In this period the column 'weeks' means the days the guest stayed inside the week price

#### Clean 'winter', 'pre_post_season' and 'high_season' columns 

```python
def clean_seasons_dates(x):
    replacements = {' jan': '/01', ' mrt': '/03', ' jun': '/06', ' sep': '/09', ' okt': '/10', ' nov': '/11', ' dec': '/12'}
    for old, new in replacements.items():
        x = x.replace(old, new)
    return x.strip()
```

```python
seasons_col_list = list(seasons.columns)
```

```python
# drop 'year' from the list 
value_to_drop = 'year'

if value_to_drop in seasons_col_list:
    seasons_col_list.remove(value_to_drop)

seasons_col_list
```

```python
# apply it to all columns but 'year' column
for col in seasons_col_list:
    seasons[col] = seasons[col].apply(clean_seasons_dates)
    
seasons
```

```python
seasons.dtypes
```

```python
# change dtypes to datetime in all columns but 'year'
for col in seasons_col_list:
    seasons[col] = pd.to_datetime(seasons[col], format='%Y/%d/%m')
```
