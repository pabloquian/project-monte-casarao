### Extract SEO information

#### Define a function that extracts Link, Title, Meta Description, h1, h2, Title length and Meta Description length of the given web pages and all inner links

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def get_page_info(url):
        # Fetch the webpage
        response = requests.get(url)
        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.text, 'html.parser')
            # Extract title and meta description of the current page
            title = soup.title.text.strip() if soup.title else "No title found"
            meta_description = get_meta_description(soup)
            # Find all anchor tags
            anchor_tags = soup.find_all('a', href=True)
            data = []
            for anchor_tag in anchor_tags:
                link = urljoin(url, anchor_tag['href'])
                title_link, meta_description_link, h1, h2 = get_link_info(link)
                data.append([link, title_link, meta_description_link, h1, h2])
            df = pd.DataFrame(data, columns=['Link', 'Title', 'Meta Description', 'h1', 'h2'])
            # Add columns with length of 'Title' and 'Meta Description'
            df['Title length'] = df['Title'].apply(text_lenght)
            df['Meta Description length'] = df['Meta Description'].apply(text_lenght)
            # Drop duplicates
            df = df.drop_duplicates(subset=['Link', 'Title', 'Meta Description', 'h1'])
            return df
        else:
            return pd.DataFrame({"Failed to retrieve webpage": [url]})

def get_meta_description(soup):
    # Find the meta description tag
    meta_tag = soup.find('meta', attrs={'name': 'description'})
    # Extract content attribute if meta tag exists
    if meta_tag:
        return meta_tag['content']
    else:
        return "No meta description found."

def get_link_info(link):
    
        # Fetch the webpage
        response = requests.get(link)
        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.text, 'html.parser')
            # Extract the title tag
            title = soup.title.text.strip() if soup.title else "No title found"
            meta_description = get_meta_description(soup)
            h1 = get_h1_from_link(soup)
            h2 = get_h2_from_link(soup)
            return title, meta_description, h1, h2
        else:
            return "Failed to retrieve webpage."

def get_h1_from_link(soup):
    # Find all h1 tags
    h1_tags = soup.find_all('h1')
    h1_texts = []
    for h1_tag in h1_tags:
        h1_texts.append(h1_tag.text.strip())
    return h1_texts[0]

def get_h2_from_link(soup):
    # Find all h2 tags
    h2_tags = soup.find_all('h2')
    h2_texts = []
    for h2_tag in h2_tags:
        h2_texts.append(h2_tag.text.strip())
    return h2_texts

def text_lenght(x):
    return len(x)

def example_usage(url_example, df_name, xlsx_name):
    url = url_example
    df_name = get_page_info(url)
    df_name.to_excel(str(xlsx_name)+'.xlsx', index=False)
    print(f'{xlsx_name} ---> has been saved')
    return df_name.head()
```

```python
# extract information from the english webpage
example_usage('https://www.montedocasarao.com/en/', 'web_seo_english', 'web_seo_english')
```
