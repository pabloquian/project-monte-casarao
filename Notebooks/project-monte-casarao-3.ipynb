In this notebook I do the following:
 - Build a model that predicts revenue for 2024 and 2025

### Read libraries

```python
import pandas as pd
import numpy as np
```

### Load dataset

```python
# I use just the bookings dataframe
bookings_cleaned_merged = pd.read_csv('bookings_cleaned_merged.csv')

pd.set_option('display.max_columns', None)
bookings_cleaned_merged
```

### extract year from the arrival column

```python
def extract_year(x):
    return x[:4]
```

```python
bookings_cleaned_merged['arrival_year'] = bookings_cleaned_merged['arrival'].apply(extract_year)
```

```python
bookings_cleaned_merged['arrival_year']
```

### aggregate final_total by year

```python
revenue_year = bookings_cleaned_merged.groupby(['arrival_year']).agg({'final_total':sum})
```

```python
revenue_year
```

```python
revenue_year = revenue_year.reset_index()
```

```python
revenue_year
```

```python
# drop the first, second and last rows
revenue_year = revenue_year.drop([0, 1, 18], axis=0).reset_index(drop=True)
```

```python
revenue_year
```

### X-y split

```python
X = revenue_year['arrival_year']
y = revenue_year['final_total']
```

```python
# Convert to array
X = X.values
y = y.values

# Reshape the array into a 2D array with one column
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

### Fit the model

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(X,y)
```

```python
# make a np array with the desired year to predict
to_predict = np.array([2024, 2025])

# Reshape the 1D array into a 2D array with one column
to_predict = to_predict.reshape(-1, 1)
```

```python
pred = lm.predict(to_predict)
```

```python
pred
```
